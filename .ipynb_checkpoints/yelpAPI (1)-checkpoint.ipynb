```python
import glob
import json

file_list = glob.glob("*.json")
file_list
```

```python
allFilesDict = {v:k for v, k in enumerate(file_list, 1)}
allFilesDict
```

```python
data = []

for k,v in allFilesDict.items():
    if 1 <= k <= 2:
        with open(v, 'r') as d:
            jdata = json.load(d)
            if jdata:
                data.append(jdata)
```
